```python
import numpy as np
import matplotlib.pyplot as plt
import math
import tqdm
```

```python
def f(x, const):
    return (x ** 2 + const)
```

```python
def g(x, const):
    return (x ** 5 + const)
```

```python
def julia_set(f, const, z, epsilon=100, max_iter=1000):
    it = 0
    while abs(z) < epsilon and it < max_iter:
        z = f(z, const)
        it += 1
    return (it / max_iter) * 2
```

```python
size=1000
X = np.linspace(-2, 2, size)
Y = np.linspace(-2j, 2j, size)
Z = np.zeros((size, size), dtype=np.complex128)

for i in range(size):
    for j in range(size):
        Z[i, j] = X[i] + Y[j]
```

```python
const = complex(-1, 1)
intensity = np.zeros((size, size))

for i in range(size):
    for j in range(size):
        intensity[i, j] = julia_set(f, const, Z[i, j])

plt.figure(figsize=(10, 10))
plt.imshow(intensity, cmap='hot')
plt.title(f'Julia Set for c = {const}')
plt.show()
```
